TRAIN TEST AYIRMA

```python
# Özellikler (X) ve hedef olarak (y) olarak data ayırma
X = data.drop(["price_range"], axis=1)
y = data["price_range"]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Model seçimi yapılıp eğitim verilerimizi modele fit ediyoruz
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

# Test datalarımızı kullanarak tahminler elde ediliyor
pred_knn = model.predict(X_test)

# Modelimizin skorlarını ve karışıklık matrisini hesaplıyoruz
knn_acc = accuracy_score(y_test, pred_knn) * 100
conf_matrix_knn = confusion_matrix(pred_knn, y_test)
report = classification_report(pred_knn, y_test)

# Sonuçlar bastırılıyor
print("Tutarlılık Oranı:" + str(format(knn_acc, ".2f")))
print("\nSınıflandırma Raporu:\n" + str(report))
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

pred_dt = model.predict(X_test)

dt_acc = accuracy_score(y_test,pred_dt) * 100
conf_matrix_dt = confusion_matrix(pred_dt,y_test)
report = classification_report(pred_dt,y_test)

print("Tutarlılık Oranı:" + str(format(dt_acc, ".2f")))
print("\nSınıflandırma Raporu:\n" + str(report))
```

NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train,y_train)

pred_nb = model.predict(X_test)

nb_acc = accuracy_score(y_test, pred_nb) * 100
conf_matrix_nb = confusion_matrix(pred_nb, y_test)
report = classification_report(pred_nb, y_test)

print("Tutarlılık Oranı:" + str(format(nb_acc, ".2f")))
print("\nSınıflandırma Raporu:\n" + str(report))
```

SVM

```python
from sklearn.svm import SVC

model = SVC(kernel = "linear", C = 1)
model.fit(X_train, y_train)

pred_svm = model.predict(X_test)

svm_acc = accuracy_score(y_test, pred_svm) * 100
conf_matrix_svm = confusion_matrix(pred_svm, y_test)
report = classification_report(pred_svm, y_test)

print("Tutarlılık Oranı:" + str(format(svm_acc, ".2f")))
print("\nSınıflandırma Raporu:\n" + str(report))
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)

pred_rf = model.predict(X_test)

rf_acc = accuracy_score(y_test, pred_rf) * 100
conf_matrix_rf = confusion_matrix(pred_rf, y_test)
report = classification_report(pred_rf, y_test)

print("Tutarlılık Oranı:" + str(format(rf_acc, ".2f")))
print("\nSınıflandırma Raporu:\n" + str(report))
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression

# Model seçimi yaparken solver olarak newton-cholesky seçilmesi büyük veri setlerinde kullanışlı olduğu için
model = LogisticRegression(solver = "newton-cholesky")
model.fit(X_train, y_train)

pred_lr = model.predict(X_test)

lr_acc = accuracy_score(y_test, pred_lr) * 100
conf_matrix_lr = confusion_matrix(pred_lr, y_test)
report = classification_report(pred_lr, y_test)

print("Tutarlılık Oranı:" + str(format(lr_acc, ".2f")))
print("\nSınıflandırma Raporu:\n" + str(report))
```
